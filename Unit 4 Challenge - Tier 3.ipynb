Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
#subset to get prices for Brent
brent = df[df['London_Borough'] == 'Brent']

#plot line graph for Month against Average Price, for Brent
brent.plot(kind='line', x='Month', y='Average_price').set_ylabel('Price')
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:
1. look through the `Month` column
2. extract the year from each individual value in that column 
3. store that corresponding year as separate column. 

Whether you go ahead with this is up to you. Just so long as you answer our initial brief: which boroughs of London have seen the greatest house price increase, on average, over the past two decades? 

```python
#extract + assign the year value from 'Month' to a new column called 'Year'
df['Year'] = df['Month'].apply(lambda t: t.year)
df.head(15)
```

```python
#now, groupby Borough and Year to get the mean of Average_price, per year/borough
df_annual_avg = df.groupby(['London_Borough','Year']).mean()
df_annual_avg = df_annual_avg.reset_index()
df_annual_avg.sample(20)
```

```python
#remove rows that are NOT 1998 and 2018 since the question at hand is only referring the change 
#between these two dates
valid_years = [1998,2018]
df_final = df_annual_avg[df_annual_avg['Year'].isin(valid_years)]
df_final.head(20)
```

```python
#test calculate a ratio for Brent
float(df_final[(df_final['London_Borough'] == 'Brent') & (df_final['Year'] == 2018)]['Average_price']) / float(df_final[(df_final['London_Borough'] == 'Brent') & (df_final['Year'] == 1998)]['Average_price'])
```

```python
#create dictionary and store ratio value for each borough
final = {b:float(df_final[(df_final['London_Borough'] == b) & (df_final['Year'] == 2018)]['Average_price']) / float(df_final[(df_final['London_Borough'] == b) & (df_final['Year'] == 1998)]['Average_price']) for b in df_final['London_Borough'].unique()}
print(final)
```

```python
#convert dictionary into dataframe called df_plot for plotting/visualization
df_plot = pd.DataFrame([final])
df_plot.head()
```

```python
#transpose df_plot now + reset index
# All we need to do now is transpose it, and reset the index! 
df_plot_T = df_plot.T
df_plot_T = df_plot_T.reset_index()
df_plot_T.head()
```

```python
# Let's just rename the 'index' column as 'London_Borough', and the '0' column to '2018|1998 price ratio'.
df_plot_T.rename(columns={'index':'Borough', 0:'price_ratio'}, inplace=True)
df_plot_T.head()
```

```python
# Sort ratios in descending order
df_plot_T = df_plot_T.sort_values('price_ratio', ascending=False)
df_plot_T
```
